```python
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from unicodedata import normalize
import requests
global pages
```

```python
my_URL = "https://www.espn.com/nba/team/stats/_/name/atl"
html = requests.get(my_URL).text
soup = BeautifulSoup(html, "html.parser")

team_names=[]
for i in soup.find_all("div",attrs={"dropdown dropdown--md h-100 pageHeading__team-stats-dropdown bp-mobile"}):
    for k in i.find_all("option"):
        team_names.append(k.get("data-param-value"))
        
team_names.remove(team_names[0])  
team_names.remove(team_names[len(team_names)-1]) 
team_names.append('atl')

print(team_names)





```

```python
year=[]
real=[]

for k in soup.find_all("option",attrs={"dropdown__option"}):
    real.append(k.get_text())
    if(k.get_text().find("20")):
        year.append(k.get_text())
        
season_year=set(real)-set(year)
season_year1=[]
for p in season_year:
    season_year1.append(p[0:7])
    
season_year1.sort()
season_year1 = list(dict.fromkeys(season_year1))
season_year1=season_year1[len(season_year1)-7:len(season_year1)]
season_year1.reverse()
print(season_year1)
```

```python
cols=[]
for k in soup.find_all("th"):
    cols.append(k.get_text())

cols = list(dict.fromkeys(cols))
print(cols)
```

```python
df = pd.DataFrame(columns=cols)
df['Season']=[]
index2=0

for team in team_names:
    for year in season_year1:
        curr_year='20'+year[5:7]
        my_URL = "https://www.espn.com/nba/team/stats/_/name/"+team+"/season/"+curr_year+"/seasontype/2"
        html = requests.get(my_URL).text
        soup = BeautifulSoup(html, "html.parser")
        print(my_URL)
        names=[]
        for k in soup.find_all("td",attrs={"class":"Table__TD"}):
            for i in k.find_all("a",attrs={"class":"AnchorLink"}):
                names.append(i.get_text())

        names = list(dict.fromkeys(names))

        att=[]
        for j in soup.find_all("table",attrs={"class":"Table Table--align-right"}):
            for k in j.find_all("tr",attrs={"class":"Table__TR Table__TR--sm Table__even"}):
                for i in k.find_all("td",attrs={"class":"Table__TD"}):
                    att.append(i.get_text())
                    
        size=len(names)*14
        att1=att[0:size]
        att2=att[size+14:len(att)-14]

        length=len(cols)-1
        start=0
        index=len(names)+1
        for i in names: 
            list1=[]
            list1.append(i)
            for k in att[start:start+14]:
                list1.append(k)
            for j in att[index*14:(index+1)*14]:
                list1.append(j)
            
            list1.append(year)

            start+=14
            index+=1
            df.loc[index2]=list1
            index2+=1
            
            
```
